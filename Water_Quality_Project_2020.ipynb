Problem statement

Groundwater is vital for the health and well-being of crops, livestock, and humans. Suboptimal groundwater quality can have devastating consequences, including crop failure and disease in livestock. Accurate knowledge of groundwater quality empowers farmers to optimize its utilization, cultivating resilient crops tailored to thrive in specific water conditions.

Classification of groundwater based on RSC

RSC is defined as the excess of carbonate and bicarbonate amount over the alkaline
earths

Permissible limit is >2.5

Due to this,the water in the soil gets more concentrated,and it declines the soil permeability.
Here, concentrations are in meq/L.


RSC less than 1.25 is safe

RSC between 1.25 and 2.50 is marginal

RSC greater than 2.50 is unsuitable

The columns in the dataset can be used to predict water quality, which are broadly classified into 9 classes-
 C1S1,C2S1,C3S1,C3S2,C3S3 ,C4S1,C4S2,C4S3,C4S4.

C1S1 to C4S4

- C refers to the salinity level (C1: low, C2: medium, C3: high, C4: very high)
- S refers to the sodium level (S1: low, S2: medium, S3: high, S4: very high)

Each classification has specific guidelines for:

- Soil type and permeability
- Crop selection (salt tolerance)
- Drainage and leaching requirements
- Organic matter additions
- Gypsum amendments

# ML Classification Model

```python
x=df[['gwl','pH','E.C','TDS','T.H','SAR']]
y=df[['RSC  meq  / L']] #independent and dependent variables
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV #for hyperparameter tuning
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=param_grid, cv=5, scoring='accuracy') #creating a grid search object
grid_search.fit(x, y) #fitting the grid search object to the data
```

```python
grid_search.best_params_ #best parameters for the model
```

```python
accuracy=grid_search.best_score_ #best accuracy
```

```python
accuracy
```
